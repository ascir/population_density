Finding the Population Density

```python
%matplotlib inline
import matplotlib.pyplot as plt
df1.plot(kind="scatter", x="longd", y="latd")
```

Increasing transparency in order to obtain a better view of the population density on the scatterplot

```python
pop= df1.plot(kind="scatter", x="longd", y="latd", title= 'population density', alpha=0.2)
pop.set_xlabel("longitude")
pop.set_ylabel("latitude")
```

Categorizing the plot using colormap to diffrentiate between areas of different populations. Red marks the region of highest population and blue for lower population densities.

```python
df1.plot(kind="scatter", x="longd", y="latd",
     s=df1['population_total']/1000, label="population",
     c="population_total", cmap=plt.get_cmap("jet"),
     colorbar=True, alpha=0.4, figsize=(10,7),
)
plt.legend()
# save_fig("housing_prices_scatterplot")
plt.show()
```
